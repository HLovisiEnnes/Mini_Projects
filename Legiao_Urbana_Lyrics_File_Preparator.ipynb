My intetion with this notebook is to create a CSV profile of Legião Urbana's (Brazilian 80's and 90's pop rock band) lyrics, which I will later use in other projects. For that, this will end up being an exercise on Beautiful Soup.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
all_songs = requests.get("https://www.letras.mus.br/legiao-urbana/") #site with links leading to 
                                                                    #all songs by the band
```

```python
content = all_songs.content
soup_all_songs =  BeautifulSoup(content)
songs = soup_all_songs.find_all("a", "song-name") #all songs by the band with links
songs
```

```python
songs[0].text
```

Now we make the links that are directly requesteable

```python
links = []
for song in range(len(songs)):
    links.append((songs[song].text, "https://www.letras.mus.br"+songs[song].get("href")))
links
```

That is, the number of songs is 

```python
len(links)
```

Not much, indeed. Let us make the list of lists for the [title, lyric]

```python
list_of_lyrics = []

for song in range(len(links)):
    link = links[song][1] #take the link of the song
    request = requests.get(link) #request it
    c = request.content
    
    soup = BeautifulSoup(c)
    samples = soup.find_all("p") #isolate the part with actual lyrics
    
    lyrics = links[song][0] #take the title of the song and add it to the lyrics
    
    for part in range(len(samples)-3): #do not consider the last 3 parts of the lyrics, as they are website specific such as
                                        #"follow this artist", etc
        lyrics += "\n\n"
        lyrics += "\n".join(samples[part].strings)
    
    list_of_lyrics.append([links[song][0], lyrics]) #make the list of titles and lyrics
```

```python
print(list_of_lyrics[13][1])
```

Faboulous. Let us only make a pandas dataframe and export the data as CSV

```python
data = pd.DataFrame(list_of_lyrics, columns = ["Title", "Lyrics"])
data
```

And finally saving it (I will do both zipped and unzipped, just in case I need either)

```python
data.to_csv("Lyrics_Legiao_Urbana.csv", index=False)
```

```python
compression_opts = dict(method='zip',
                        archive_name='lyrics_legiao_urbana.csv')  
data.to_csv('lyrics_legiao_urbana.zip', index=False,
          compression=compression_opts)  
```
