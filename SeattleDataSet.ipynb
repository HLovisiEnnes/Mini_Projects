```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw_data = pd.read_csv("seattleWeather_1948-2017.csv")
```

```python
raw_data
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data = raw_data.drop("DATE",axis=1)
data["TAVG"]=(data["TMAX"]+data["TMIN"])/2
data
```

```python
data = scaler.fit_transform(data)
data
```

```python
temp_min = data[:,2]
temp_max = data[:,1]
temp_avg = data[:,-1]
perception = data[0]
```

```python
np.save("Temperature_min",temp_min)
np.save("Temperature_max",temp_max)
np.save("Temperature_avg",temp_avg)
np.save("Perception",perception)
```
